```python
def number_sqaured(number):
    print(number**2)
```

```python
number_sqaured(5)
```

```python
def number_sqaured_custom(number, power):
    print(number**power)
```

```python
number_sqaured_custom(5, 3)
```

```python
#Arbitrary Arguments: you can specify how many arguments you want to pass to the function later

def number_args(*number):
    print(number[0]*number[1])
```

```python
number_args(5,6,1,2,8)
```

```python
args_tuple = (5,6,1,2,8)

def number_args(*number):
    print(number[0]*number[1])
    
number_args(*args_tuple)
```

Key Word Arguments

```python
def number_sqaured_cust(number, power):
    print(number**power)
```

```python
number_sqaured_cust(power = 5, number = 3)
```

```python
#Artbitrary keyword argument

def number_kwarg(**number):
    print('My number is: ' + number['integer'] + ' My other number: ' + number['integer2'])
```

```python
number_kwarg(integer = '2309', integer2 = '349')
```
