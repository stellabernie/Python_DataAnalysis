For Loops

```python
#Used to iterate over a sequence: eg. list, tuple, array, dictionary
```

```python
integers = [1,2,3,4,5]
```

```python
for number in integers:
    print (number)
```

```python
for number in integers:
    print ("yep!")
```

```python
integers = [1,2,3,4,5]
```

```python
#in this case, the parenthesis speaks to adding the variables to each other as they are looped
for Jelly in integers:
    print(Jelly + Jelly)
```

```python
#Applied to Dictionaries
```

```python
ice_cream_dict = {'name': 'Alex Freberg', 'weekly intake':5, 'favourite ice creams':['ChocChip', 'Salted Vanilla']}
```

```python
for cream in ice_cream_dict.values():
    print(cream)
```

```python
for key, value in ice_cream_dict.items():
    print(key, '->', value)
```

Nested For Loops

```python
flavors = ['Vanilla', 'Chocolate', 'Cookie Dough']
toppings = ['Hot Fudge', 'Oreos', 'Marshmallows']
```

```python
for one in flavors: 
    for two in toppings:
        print(one, 'topped with', two)
```
