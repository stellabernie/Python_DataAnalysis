# Scraping Data From a Website

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_French_companies'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
#Note, if the outcome above was all the details of the tables on the page, you could extract the table you want. You can do this by using indexing, as the information is returned as a lit (all tables are separated by comma's), you can pull each table by specifying the position in this grand list. 
#eg.  soup.find_all('table')[1]
#eaxmple below
```

```python
soup.find_all('table')
```

```python
soup.find_all('table')[0]
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
table.find_all('th')
```

```python
world_titles = table.find_all('th')
```

```python
print(world_titles)
```

```python
world_table_titles = [title.text.strip() for title in world_titles]

#Note 'strip' helps you to clean it
```

```python
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
```

```python
df
```

```python
#Used the inspect tool to find out what 'class?' is used for row data since we have column data above. Identified it by 'tr' numbered for each row.


table.find_all('tr')
```

```python
#our row of data
column_data = table.find_all('tr')
```

```python
#we are looping through our row data to look at individual data tags. Getting out the text, and stripping it to clean it.
#If we just print 'column_data' just like that - then we encounter an issue bc the first row is blank. So to make sure we start off at the second column
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    #print(individual_row_data)
    
    
#Then we are looking at current dataframe,which has nothing in it at present, we're looking at the length of it and we are appending each row of the information into the next posistion
    length = len(df)
    df.loc[length] = individual_row_data
```
