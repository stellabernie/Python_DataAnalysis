If - Elif - Else Statements

```python
#You can have many elif statements but can only have 1 if and 1 else.
#<,>,<=,>= etc are comaparison operators
#and,or etc are logical operators
```

```python
if 25>10:
    print('It worked!')
```

```python
if 25<10:
    print('It worked!')
else: 
    print ('It did not work...')
```

```python
if 25>10:
    print('It worked!')
else: 
    print ('It did not work...')
```

```python
if 25<10:
    print('It worked!')
elif 25< 30:
    print('elif worked!')
else: 
    print ('It did not work...')
```

```python
if 25<10:
    print('It worked!')
elif 25< 20:
    print('elif worked!')
elif 25< 21:
    print('elif 2 worked!')  
elif 25< 40:
    print('elif 3 worked!')
elif 25< 50:
    print('elif 4 worked!')
else: 
    print ('It did not work...')
```

```python
if (25 < 10) or (1 < 3):
    print('It worked!')
elif 25<10:
    print('It worked!')
elif 25< 20:
    print('elif worked!')
elif 25< 21:
    print('elif 2 worked!')  
elif 25< 40:
    print('elif 3 worked!')
elif 25< 50:
    print('elif 4 worked!')
else: 
    print ('It did not work...')
```

```python
print ('It worked!') if 10>30 else print ('It did not work.....')
```
