While Loop

```python
#For Loop will iterate over the entire sequence regardless of a condition , While Loop will iterate over a sequence only if that specific condition is met
```

```python
number = 0
while number < 5:
    print(number)
    number = number + 1 
```

```python
#Break statment: we can stop the loop even if the while condition is true

number = 0

while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1 
```

```python
#else statement will only execute when the while statment is no longer true

number = 0

while number < 5:
    print(number)
    if number == 6:
        break
    number = number + 1
else:
    print('No longer < 5')
```

```python
number = 0

while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
else:
    print('No longer < 5')
```

```python
#Continue Statement: when the continue statement is triggered, it rejects all remaining statements in the current iteration of the loop then will go to the next iteration
```

```python
#...This creates an infinte loop

number = 0

while number < 5:
    print(number)
    if number == 3:
        continue
    number = number + 1
else:
    print('No longer < 5')
```
